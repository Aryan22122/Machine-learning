```python
def fn2(X,y,b,m):
    n = X.shape[0]
    
    slope = 0
    m_overall=0
    for i in range(n):
        slope = slope + y[i] - m*X[i] - b
        m_overall = m_overall + (y[i] - m*X[i]-b)*X[i]


    return -2*slope,-2*m_overall
    
```
