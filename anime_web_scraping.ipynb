```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
def extractor(soup,i):   
    text = soup.find_all('h3')[i].text
    clean = re.sub(r'^\d+\.\s*', '', text)
    overall_list = []
    overall_list.append(clean)
    
    review = soup.find_all('div',class_="information di-ib mt4")[i].text.strip()
    list_of_things = [line.strip() for line in review.split("\n") if line.strip()]
    overall_list.extend(list_of_things)
    imdb = soup.find_all("div",class_="js-top-ranking-score-col top-ranking-score-col di-ib al")[i].text
    overall_list.append(imdb)
    # print(overall_list)
    return overall_list

```

```python
df = pd.DataFrame(columns=['name','no_of_episodes','release_data','no_of_members','imdb'])
```

```python
from tqdm import tqdm
```

```python
for i in tqdm(range(0,29351,50),desc="prgress->"):
    webpage = requests.get(f'https://myanimelist.net/topanime.php?limit={i}',
                           headers={'User-Agent':'Mozilla/5.0'})
   
    soup = BeautifulSoup(webpage.text,'lxml')
    for j in range(0,50):
        overall_list = extractor(soup,j)
        df.loc[len(df)] = overall_list
    
    
```
