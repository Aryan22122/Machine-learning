```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=100,n_features=2,n_informative=1,hypercube=False,random_state=41,n_clusters_per_class=1,n_redundant=0,n_classes=2,class_sep=30)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0],X[:,1],c = y,cmap='winter')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(penalty=None,solver='saga')
```

```python
lr.fit(X,y)
```

```python
coef_ = lr.coef_
```

```python
coef_
```

```python
intercept_ = lr.intercept_
```

```python
intercept_
```

```python
# Ax + By + C = 0
# y = -A/B * x - C/B 

# y = mx+c

m = -(coef_[0][0]/coef_[0][1])
c = -(intercept_/coef_[0][1])
```

```python
m  
```

```python
c
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c
```

```python
plt.plot(x_input,y_input,color='blue',linewidth=3)
plt.scatter(X[:,0],X[:,1],c = y,cmap='winter')
plt.ylim(-3,2)
```

```python
class myLR:
    def __init__(self,lr=0.01,epochs=1000):
        self.lr = lr
        self.epochs=epochs
        self.coef_ = None
        self.intercept_ = None
    def fit(self,X,y):
        X = np.insert(X,0,1,axis=1)
        W = np.ones(X.shape[1]) 
        for i in range(self.epochs):
            
            y_pred = step(np.dot(X,W))

            W = W + self.lr*(y - y_pred).dot(X)
        self.coef_ = W[1:]
        self.intercept_ = W[0]
        return self.coef_,self.intercept_
            
            
    
    
```

```python
lr = myLR(0.001,3000)
    
```
