```python
from sklearn.metrics import r2_score
```

```python
import random
```

```python
class MBGDregressor:
    def __init__(self,batch_size,learning_rate = 0.1,epochs=100):
        self.learning_rate = learning_rate
        self.epochs= epochs
        self.intercept_ = None
        self.coef_ = None
        self.batch_size = batch_size
    def fit(self,X_train,y_train):
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])

        
        for i in range(self.epochs):
            for j in range(int(X_train.shape[0]//self.batch_size)):
                idx = random.sample(range(X_train.shape[0]),self.batch_size)
                y_hat = np.dot(X_train[idx],self.coef_) + self.intercept_
                intercept_diff = -2*(y_train[idx]-y_hat).mean()
                coeff_diff = -2*((y_train[idx]-y_hat).dot(X_train[idx]))
        
                self.intercept_ = self.intercept_ - (self.learning_rate*intercept_diff)
                self.coef_ = self.coef_ - (self.learning_rate*coeff_diff)
        return self.coef_,self.intercept_
                
    def predict(self,X_test):
        y_pred = np.dot(X_test,self.coef_) + self.intercept_
        return y_pred
```

```python
mb = MBGDregressor(32,0.1,100)
coef_fn,inter_fn = mb.fit(X_train,y_train)
```

```python
y_pred = mb.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
